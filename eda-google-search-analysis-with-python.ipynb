```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

* Google doesnt give much access to the data about daily search queries, but another application of google known as Google Trends can be used for the task of Google search analysis.

* Google Trends provides an API that can be used to analyze the daily searches on Google. This is known as pytrends, lets install it.


```python
!pip install pytrends
```

```python
from pytrends.request import TrendReq
trends = TrendReq()
```

Here we will be analyzing the google search trends on the queries based on "Machine Learning", so lets create a DataFrame of the top 10 countries which search for "Machine Learning" on Google.

```python
trends.build_payload(kw_list=['Machine Learning'])
data = trends.interest_by_region()
data = data.sort_values(by='Machine Learning', ascending=False)
data = data.head(10)
data
```

So,according to the above results, the search queries based on "Machine Learning" are mostly done in "China". Lets create the visualisation using "bar" chart.

```python
data.index
```

Lets reset the index

```python
data.reset_index(inplace=True)
data
```

```python
data.plot(x="geoName", 
          y="Machine Learning", 
          figsize=(15,12), 
          kind='bar', 
          title='Top 10 countries which search for "Machine Learning" on Google.',
         )
plt.style.use('fivethirtyeight')
plt.show()
```

Lets analyse the trend of searches to see how the total search queries based on "Machine Learning" increased or decreased on Google.

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Machine Learning', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
